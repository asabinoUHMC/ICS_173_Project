```python
import pandas as pd 
import seaborn as sns 
import matplotlib.pyplot as plt
import numpy as np 
```

```python
df = pd.read_csv(r"S:\School\ICS 173\best-selling game consoles and their best-selling games.csv")
df.head(50)
```

```python
df_sorted = df.sort_values(by="Units Sold (in millions)", ascending=False)
df_sorted.head()
```

```python
top_n = 50
top_games = df_sorted.head(top_n)
```

```python
plt.figure(figsize=(12,8))
sns.barplot(x="Units Sold (in millions)", y="Most Sold Game", data=top_games, palette='viridis')
plt.title(f"Top {top_n} Most Sold Games")
plt.xlabel("Units Sold (in millions)")
plt.ylabel("Games")
plt.show()
```
